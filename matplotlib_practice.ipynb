```python
# Here is oo stype

x = np.linspace(0, 2, 100) # from 0 to 2, 100 points 
# create a figure and with one ax
fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')  # label here is corresponding to legend()
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple plot')
ax.grid(True, color='r')  # set the grid parameters 
ax.legend()        # add a legend
```

#### draw bar figures

```python
import numpy as np  # facilate calculate array - scalar 

plt.style.use('fivethirtyeight')  # set style to use
ages_x = np.array([i for i in range(25, 36)])
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
dev_y2 = [i-200 for i in dev_y]
width = 0.25 # define a width to seperate two bars
fig, ax = plt.subplots()
ax.bar(ages_x-width, dev_y, width=width) # how to shift? -+width 
ax.bar(ages_x, dev_y2, width=width ) # tell the bar the width
```

```python
# draw a horizontal bar chart
ages_x = np.array([i for i in range(25, 36)])
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
fig, ax = plt.subplots()
ax.barh(ages_x, dev_y)  # just change bar to barh() 
```

#### draw pie chart

```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight') # use fivethirtyeight style sheet
```

```python
plt.title('my awesome pie chart')
slices = [60, 40] # pie chart is very straightforward
labels = ['sixty', 'forty'] # label list
colors = ['red', 'green'] # use a color list
explode=[0, 0.1] # shadow=True, it will make the pie chart look more 3D
plt.pie(slices, colors=colors ,labels=labels, explode=explode,shadow=True,wedgeprops={'edgecolor':'black'})
plt.tight_layout()
plt.show()
```

#### scatter plot

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
```

```python
plt.style.use('seaborn')
```

```python
plt.scatter(x, y, s=100, c=colors, cmap='Greens',
           edgecolor='black', linewidth=1, alpha=0.75) # s parameter means the size of dot


cbar = plt.colorbar() # set color bar 
cbar.set_label('Satisfaction')
plt.xscale('log')
plt.ylabel('log') # we could set xscale to log scale 
```

#### time series data

```python
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('seaborn')
```

```python
dates = [
    datetime(2019, 5, 24), 
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29), 
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()  # gcf means get current figure
```
