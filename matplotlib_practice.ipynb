```python
# two ways to create figure, which matplotlib graphs your data
figure1 = plt.figure() # the simple way
figure2, ax = plt.subplots() # figure with one axes
fig, axs = plt.subplots(2, 2) # a figure with a 2x2 grid of Axes
```

```python
# Here is oo stype

x = np.linspace(0, 2, 100) # from 0 to 2, 100 points 
# create a figure and with one ax
fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')  # label here is corresponding to legend()
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple plot')
ax.grid(True, color='r')  # set the grid parameters 
ax.legend()        # add a legend
```

#### draw bar figures

```python
import numpy as np  # facilate calculate array - scalar 

plt.style.use('fivethirtyeight')  # set style to use
ages_x = np.array([i for i in range(25, 36)])
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
dev_y2 = [i-200 for i in dev_y]
width = 0.25 # define a width to seperate two bars
fig, ax = plt.subplots()
ax.bar(ages_x-width, dev_y, width=width) # how to shift? -+width 
ax.bar(ages_x, dev_y2, width=width ) # tell the bar the width
```

```python
# draw a horizontal bar chart
ages_x = np.array([i for i in range(25, 36)])
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
fig, ax = plt.subplots()
ax.barh(ages_x, dev_y)  # just change bar to barh() 
```
